##  How can you convert the 'OrderDateKey' column from integer to datetime data type using pandas?

```python

df['OrderDateKey'] = pd.to_datetime(df['OrderDateKey'], format='%Y%m%d')
```

## In the 'Unit Price' column, there are currency values stored as strings. How can you convert them to floating-point numbers?


```python
df['Unit Price']=df['Unit Price'].astype('float')
```

## The 'Sales Order' column contains alphanumeric order IDs. How can you convert them to a categorical data type?


```python
df["Sales Order"]=df["Sales Order"].astype('category')
```

# The 'Order Quantity' column is currently stored as a float. How can you convert it to an integer data type?


```python
df["Order Quantity"]=df["Order Quantity"].astype('int')
```

# The 'Color' column contains color names as strings. How can you convert them to a categorical data type?


```python
df["Color"]=df["Color"].astype('category')
```
